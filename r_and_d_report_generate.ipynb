```python
def generate_prompt(title,summary):
    prompt = f"""
あなたは、優秀な法律関連の出版社の研究開発部(R&D)です。以下の研究報告書の概要に基づいて、研究報告書を具体的な内容を作成してください。
タイトル:{title}
研究報告書の概要:
{summary}

研究報告書では、以下の点を必ず含めてください:
1. 研究の背景と目的
2. 研究方法
3. 研究結果
4. 考察と今後の展望

研究報告書:
"""
    return prompt.strip()
```

```python
for topic in topics:
    prompt = generate_prompt(topic["title"],topic["summary"])

    messages = [
        {
            "role": "user",
            "content": prompt,
        }
    ]
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=1,
        max_tokens=2000
    )

        # 現在の日時を取得
    now = datetime.datetime.now()
    # 日時を文字列に変換し、ファイル名に使用する形式にします（例: '2023-03-27_15-30-00'）
    formatted_time = now.strftime('%Y-%m-%d_%H-%M-%S')
    # ファイル名に日時とパラメータを含める
    filename = f'./{folder_name}/response_{formatted_time}.txt'
    # テキストファイルに結果を書き込む
    with open(filename, 'w') as file:
        file.write(response.choices[0].message.content)
```

```python
import zipfile

def zip_directory(directory_path, output_path):
    with zipfile.ZipFile(output_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(directory_path):
            for file in files:
                file_path = os.path.join(root, file)
                zipf.write(file_path, os.path.relpath(file_path, directory_path))

    print(f"ディレクトリ {directory_path} が {output_path} にzipファイルとして保存されました。")

# 使用例
directory_to_zip = f"./{folder_name}"  # zipファイルにするディレクトリのパス
output_zip_file = f"./{folder_name}.zip"  # 出力するzipファイルのパス

zip_directory(directory_to_zip, output_zip_file)
```
